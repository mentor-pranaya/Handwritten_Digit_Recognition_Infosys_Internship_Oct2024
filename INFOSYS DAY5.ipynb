```python
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader


```

```python
# Define the transformation pipeline
transform = transforms.Compose([
    transforms.RandomRotation(10),            # Step 3: Random rotation for data augmentation
    transforms.RandomHorizontalFlip(),        # Step 3: Random horizontal flip for data augmentation
    transforms.ToTensor(),                    # Step 4: Convert to tensor
    transforms.Normalize((0.5,), (0.5,))      # Step 5: Normalize pixel values to [-1, 1]
])

```

```python
# Load the MNIST dataset with the transformations applied
dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)

# Create a DataLoader with batching and shuffling
dataloader = DataLoader(dataset, batch_size=10, shuffle=True)  # Steps 6 & 7: Batch and shuffle

```

```python
# Get a batch of images and labels
data_iter = iter(dataloader)
images, labels = next(data_iter)

```

```python
# Function to display images
def show_images(images, labels, num_images=10):
    import matplotlib.pyplot as plt
    num_cols = 5
    num_rows = num_images // num_cols
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(28, 28))
    axes = axes.flatten()
    for i in range(num_images):
        ax = axes[i]
        ax.imshow(images[i].squeeze(), cmap='gray')
        ax.set_title(f'Label: {labels[i]}')
        ax.axis('off')
    plt.tight_layout()
    plt.show()

```

```python
# Show the preprocessed images
show_images(images, labels, num_images=10)
```
